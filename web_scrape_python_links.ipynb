```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time


options = Options()
options.add_argument("--headless=new")
options.add_argument("start-maximized");
options.add_argument("disable-infobars");
options.add_argument("--disable-extensions");
options.add_argument("--disable-gpu");
options.add_argument("--disable-dev-shm-usage");
options.add_argument("--no-sandbox");
driver = webdriver.Chrome(options);

driver.get("https://www.ecb.europa.eu/press/accounts/html/index.en.html")
links_container = driver.find_element(By.XPATH, "//dl[@id='lazyload-container']").find_elements(By.XPATH, "div")

height = driver.execute_script("return document.body.scrollHeight")
intervals = 10
height_increment = height / intervals
for i in range(intervals):
  new_height = (i+1) * height_increment
  driver.execute_script(f'window.scrollTo(0, {new_height});')
  time.sleep(2)
links = []
for element in links_container:
  dds = element.find_elements(By.XPATH, "dd")
  for dd in dds:
    title = dd.find_element(By.CLASS_NAME, "title")
    links.append(title.find_element(By.TAG_NAME, "a").get_attribute("href"))

```

```python
links
driver.quit()
```

```python
import pandas as pd

links_df = pd.DataFrame(links, columns=['Link'])

# Save DataFrame to CSV file
links_df.to_csv('links.csv', index=False)
```
